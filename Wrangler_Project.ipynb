# Bag 6 : Mapping Kategori

```python
df2['experience_level'].unique()
```



*   SE = SENIOR
*   MI = MIDDLE
*   EN = ENTRY
*   EX = EXECUTIVE



dictionary = type data di python

```python
dict_exp_level = {
    'SE' : 'SENIOR',
    'MI' : 'MIDDLE',
    'EN' : 'ENTRY',
    'EX' : 'EXECUTIVE'
}
```

```python
df2['experience_level'] = (df2['experience_level'].map(dict_exp_level))
```

```python
df2['experience_level'].unique()
```

```python
dict_exp_level = {
    'FT' : 'Full Time',
    'CT' : 'Contract',
    'FL' : 'Freelance',
    'PT' : 'Part Time'
}
```

```python
df2['employment_type'] = (df2['employment_type'].map(dict_exp_level))
```

```python
df2['employment_type'].unique()
```

```python
df2['remote_ratio'].unique()
```

ratio 0 = Onsite
ratio 50 = Hybrid
ratio 100 = WFH

```python
dict_remote_ratio = {
    '0' : 'ONSITE',
    '50' : 'HYBRID',
    '100' : 'REMOTE'
}
```

bikin kolom kategori_remote

```python
df2['remote_category'] = (df2['remote_ratio'].map(dict_remote_ratio))
```

# Bag 8 : Grouping Data

pivot table ( ga ada jadi kolom )

berapakah rata-rata gaji berdasarkan exprience_level?

```python
df2.columns
```

```python
df2.groupby('experience_level')['salary'].mean()
```

tampilkan data maksimum dari salary berdasarkan employment_type

```python
df2.groupby('employment_type')['salary'].max()
```

```python
df2.groupby(['experience_level', 'employment_type'])[['salary', 'salary_in_usd']].mean()
```

```python
df2.groupby(['company_size', 'remote_ratio'])[['salary', 'salary_in_usd']].max().reset_index()
```

```python
df2.columns
```

```python
df2.groupby('experience_level').agg({
    'salary' : ['max', 'min', 'mean'],
    'work_year' : 'mean'
})
```

```python
df2.groupby('job_title').agg({
    'salary' : ['sum', 'min', 'mean'],
    'experience_level' : 'size'
}).reset_index()
```
